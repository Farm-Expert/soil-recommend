```python
import pandas as pd

df=pd.read_csv("/content/drive/MyDrive/crop_data.csv")

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df.STATE=le.fit_transform(df.STATE)
df.SOIL_TYPE=le.fit_transform(df.SOIL_TYPE)

label_dict={}
for label, encoded_value in zip(le.classes_, le.transform(le.classes_)):
    label_dict[label] = encoded_value
print(label_dict)
```

```python
data=df.drop(["SOIL_TYPE","CROP_PRICE","CROP"],axis=1)

target=df.drop(["CROP_PRICE","CROP"],axis=1)
print(data.shape)
print(target.shape)
```

```python
df.SOIL_TYPE.unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(data,target, test_size=0.2, random_state=42)

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor

model=MultiOutputRegressor(RandomForestRegressor(n_estimators=100))

model.fit(x_train,y_train)
```

```python
soil_pred=model.predict(x_test)
model.predict([[0,	90,	42,	43,	20.879744,	82.002744,	6.502985,	202.935536]])
```

```python
import joblib

joblib.dump(model,"soil_prediction")
```

```python
soil=joblib.load("soil_prediction")

data_input=[  0.  ,  90.        ,  42.        ,
         43.        ,  20.87595224,  81.99636968,   6.50277946,
        203.18594832]

output=soil.predict([data_input])

print(output)
```
